### 4. Exploring Data ( Visualization )

```python
plt.figure(figsize=(12,5))
sns.countplot(x=df['habitat'], data=df, hue='class', palette='dark')
plt.xlabel('Habitat')
plt.ylabel('Number of Mushroom')
plt.title('Poisionous and Palatable distribution of different habitat')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(12,5))
sns.stripplot(x=df['population'], y=df['habitat'], marker='o', color='m', size=6, edgecolor='white',orient='v')
plt.xlabel('Population')
plt.ylabel('Habitat')
plt.title('')
plt.legend()
plt.show()
```
